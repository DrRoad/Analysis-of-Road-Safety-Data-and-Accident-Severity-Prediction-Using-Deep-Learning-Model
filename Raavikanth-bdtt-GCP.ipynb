## 5.	BIG DATA, CLOUD COMPUTING AND GOOGLE CLOUD PLATFORM<a class="anchor" id="five"></a>
#### 5.1	What is a Big Data?
"Big data" is a field that treats ways to analyse, systematically extract information from, or otherwise deal with data sets that are too large or complex to be dealt with by traditional data-processing application software. Data with many cases (rows) offer greater statistical power, while data with higher complexity (more attributes or columns) may lead to a higher false discovery rate. Big data challenges include capturing data, data storage, data analysis, search, sharing, transfer, visualization, querying, updating, information privacy and data source. Big data was originally associated with three key concepts: volume, variety, and velocity. 

#### 5.2.	What is a Cloud Computing?
Cloud computing is the on demand availability of computer system resources, especially data storage and computing power, without direct active management by the user. The term is generally used to describe data centers available to many users over the Internet. Large clouds, predominant today, often have functions distributed over multiple locations from central servers. If the connection to the user is relatively close, it may be designated an edge server.

##### Service Models
Though service-oriented architecture advocates "everything as a service" (with the acronyms EaaS or XaaS, or simply aas), cloud-computing providers offer their "services" according to different models, of which the three standard models per NIST are Infrastructure as a Service (IaaS), Platform as a Service (PaaS), and Software as a Service (SaaS). These models offer increasing abstraction; they are thus often portrayed as a layers in a stack: infrastructure-, platform- and software-as-a-service, but these need not be related. For example, one can provide SaaS implemented on physical machines (bare metal), without using underlying PaaS or IaaS layers, and conversely one can run a program on IaaS and access it directly, without wrapping it as SaaS.
##### SaaS: Software as a Service
Software as a Service, also known as cloud application, services, represents the most commonly utilized option for businesses in the cloud market. SaaS utilizes the internet to deliver applications, which are managed by a third-party vendor, to its users. A majority of SaaS applications are run directly through the web browser, and do not require any downloads or installations on the client side.
Examples of SaaS
Google Cloud SQL, AWS EC2, AWS , Salesforce, Cisco WebEx, Concur, GoToMeeting
##### PaaS: Platform as a Service
Platform as a Service (PaaS) or Application Platform as a Service (aPaaS) or platform-based service is a category of cloud computing services that provides a platform allowing customers to develop, run, and manage applications without the complexity of building and maintaining the infrastructure typically associated with developing and launching an app.
##### IaaS: Infrastructure as a Service
Infrastructure as a service (IaaS) are online services that provide high-level APIs used to dereference various low-level details of underlying network infrastructure like physical computing resources, location, data partitioning, scaling, security, backup etc. A hypervisor, such as Xen, Oracle VirtualBox, Oracle VM, KVM, VMware ESX/ESXi, or Hyper-V, LXD, runs the virtual machines as guests. Pools of hypervisors within the cloud operational system can support large numbers of virtual machines and the ability to scale services up and down according to customers' varying requirements.

Alongside a set of management tools, it provides a series of modular cloud services including computing, data storage, data analytics and machine learning. ### 10.2.3.Traffic Accident Casualties by Time of Day

### 11.1.2. Data Preprocessing and Feature Creation/Selection

#### Avoiding Dummy Variable Trap 
The Dummy variable trap is a scenario where there are attributes which are highly correlated (Multicollinear) and one variable predicts the value of others. When we use one hot encoding/keras.to_categorial for handling the categorical data, then one dummy variable (attribute) can be predicted with the help of other dummy variables. Hence, one dummy variable is highly correlated with other dummy variables. Using all dummy variables for regression models lead to dummy variable trap. So, the classifier models should be designed excluding one dummy variable.

```python
from keras.utils import to_categorical
X = total_df.iloc[10000:20000,[5,10,12,14,16,17,20,24,25,26,27,33,43,44,45,46,47,48,49,50]].values
y = total_df.iloc[10000:20000,6].values
# Avoiding Dummy Varaible Trap
X = pd.concat([pd.DataFrame(pd.DataFrame(to_categorical(X[:,0])).values[:,1:]), 
                    pd.DataFrame(pd.DataFrame(to_categorical(X[:,1])).values[:,1:]), 
                    pd.DataFrame(pd.DataFrame(to_categorical(X[:,2])).values[:,1:]), 
                    pd.DataFrame(pd.DataFrame(to_categorical(X[:,3])).values[:,1:]), 
                    pd.DataFrame(pd.DataFrame(to_categorical(X[:,4])).values[:,1:]), 
                    pd.DataFrame(pd.DataFrame(to_categorical(X[:,6])).values[:,1:]), 
                    pd.DataFrame(pd.DataFrame(to_categorical(X[:,7])).values[:,1:]), 
                    pd.DataFrame(pd.DataFrame(to_categorical(X[:,8])).values[:,1:]), 
                    pd.DataFrame(pd.DataFrame(to_categorical(X[:,9])).values[:,1:]), 
                    pd.DataFrame(pd.DataFrame(to_categorical(X[:,10])).values[:,1:]), 
                    pd.DataFrame(pd.DataFrame(to_categorical(X[:,11])).values[:,1:]), 
                    pd.DataFrame(pd.DataFrame(to_categorical(X[:,12])).values[:,1:]), 
                    pd.DataFrame(pd.DataFrame(to_categorical(X[:,13])).values[:,1:]), 
                    pd.DataFrame(pd.DataFrame(to_categorical(X[:,14])).values[:,1:]),
                    pd.DataFrame(pd.DataFrame(to_categorical(X[:,16])).values[:,1:]), 
                    pd.DataFrame(pd.DataFrame(to_categorical(X[:,18])).values[:,1:]), 
                    pd.DataFrame(pd.DataFrame(to_categorical(X[:,19])).values[:,1:])], axis=1,ignore_index=True).values
```

### 11.1.3. Feature scalling

### 11.1.4. Train ANN model using Keras Deep learning framework (Evaluating, Improving and Hyperparameter Tuning of ANN Model)
ANN has two types of parameters
1. Parameters that are learnt from model (weights)
2. Hyperparameters that are stay fixed during entire process (which need to be tuned for improving model accuracy)

#### 9.1.4.1. Hyperparameter Tuning

The hyper-parameters of the ANN model were selected by performing a systematic grid searchimplemented in scikit-learn using 10, 50 epochs. Even though the systematic grid search requireshigh computational cost, better results could be obtained by systematically tuning the hyper-parametervalues.Models with various combination of parameters were constructed, and a 3-fold cross-validationwas used to evaluate each model. The parameters of the model with the highest validation accuracywere found to be the best parameters among the evaluated ones. 

```python
# Hyperparameter tuning
def build_classifier(optimizer='adam', init='normal', dropout=0.0):
    classifier = Sequential()
    # Adding the inpurt layer and first hidden layer with dropout
    # If p value in Dropout is 0.n means n neuron will be disbled for each iteration
    # Started with 0.1 and continued up to 0.3 to avoid model overfitting
    classifier.add(Dense(output_dim=64, activation='relu', init = 'uniform', input_dim = X_train.shape[1]))
    classifier.add(Dropout(rate=dropout))
     # Adding the second hidden layer with dropout
    classifier.add(Dense(output_dim=64, activation='relu'))
    classifier.add(Dropout(rate=dropout))
    # Adding the third hidden layer with dropout
    classifier.add(Dense(output_dim=64, activation='relu'))
    classifier.add(Dropout(rate=dropout))
    classifier.add(Dense(len(np.unique(y_train)), activation='softmax'))
    classifier.compile(loss='categorical_crossentropy', optimizer= optimizer, metrics=['accuracy'])
    return classifier;

classifier = KerasClassifier(build_fn=build_classifier)
# Create a Dictionary of Hypertuning parameters
# rmsprop can be best one of the best optimizer for recurrent neural networks
sgd = SGD(lr=0.01, decay=0.8, momentum=0.8, nesterov=True)
#ANN turning with different hyperparameters
optimizers = ['adam', sgd]
epochs = [10, 50]
batches = [10, 20]
dropouts = [0.1, 0.2, 0.3]
init = ['normal', 'uniform']

parameters= dict(optimizer=optimizers, epochs=epochs, batch_size=batches, init=init, dropout = dropouts)

grid_search = GridSearchCV(estimator = classifier, 
                           param_grid= parameters,
                           scoring = 'accuracy',
                           cv =3)

grid_result = grid_search.fit(X_train, y_train)
#It will take atleast some hours to tune and to get best hyperparameter scores
best_prameter = grid_result.best_params_
best_accuracy = grid_result.best_score_
```
